```python
import pandas as pd
import csv
import seaborn as sns
import numpy as np
import networkx as nx
import graphviz
import random
import matplotlib.pyplot as plt
import itertools 

from IPython.display import display
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
from sklearn.tree import ExtraTreeClassifier, DecisionTreeClassifier, export_graphviz
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier, ExtraTreesClassifier,RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import RidgeClassifierCV
from yellowbrick.model_selection import learning_curve
from sklearn.model_selection import StratifiedKFold

class repiceModel:

  def processing():
    data = pd.read_csv('/content/drive/MyDrive/repiceModel/plats.csv')
    
    #data.describe(include = 'all')
    # Check for missing values
    #data.isnull().sum().any()
    # Check for duplicates
    #data.duplicated().any()
    multi_label_binarizer = MultiLabelBinarizer()
    treated_data = multi_label_binarizer.fit_transform(data['ingredients'].str.strip().str.split(','))
    ingredients_binarizer = pd.DataFrame(treated_data, columns=multi_label_binarizer.classes_, index=data['ingredients'].index)
    #ingredients_binarizer
    dt = pd.concat([data['noms plats'], ingredients_binarizer], axis=1)
    dt.drop_duplicates()
    dt = dt[dt.duplicated(subset=['noms plats'], keep=False)]

    #dt.info
    #dt

    return dt
    

  def visualisatoion(dt):
    #Using Pearson Correlation
    plt.figure(figsize=(75,75))
    cor = dt.corr()
    sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
    plt.show()
    

  def algo(dt):

    y = dt['noms plats'].values
    X = dt.drop('noms plats', axis = 1).values
    X_train, X_test, y_train, y_test = \
            train_test_split(X, y, test_size=.31405, random_state=25, stratify=y, shuffle=True)
   
    extra_tree = ExtraTreeClassifier(random_state=10)
    names = [ "Decision Tree", "Random Forest","Nearest Neighbors",
            " ExtraTree Classifier Bagging Classifier",
            ]

    classifiers = [DecisionTreeClassifier(),
        RandomForestClassifier(max_depth=10, random_state=0 ,
                              n_estimators=141),
        KNeighborsClassifier(n_neighbors=5),   
        BaggingClassifier(extra_tree, random_state=20)]
      
    for name, clf in zip(names, classifiers):
            
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)     
        print(score,name)

    clf = RandomForestClassifier(max_depth=10, random_state=0 ,n_estimators=141)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    print("Score RandomForestClassifier :",score)
    y_pred = clf.predict(X_test)
    return y_pred, y_test, clf,X,y

  def classification_report(y_test, y_pred):
    matrix = classification_report(y_test, y_pred)
    return matrix

  def plot_confusion_matrix(y_true,y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)

    ## Get Class Labels
    class_names = classes
    # Plot confusion matrix in a beautiful manner
    fig = plt.figure(figsize=(25, 25))
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax, fmt = 'g'); #annot=True to annotate cells
    # labels, title and ticks
    ax.set_xlabel('Predicted', fontsize=20)
    ax.xaxis.set_label_position('bottom')
    plt.xticks(rotation=90)
    ax.xaxis.set_ticklabels(class_names, fontsize = 10)
    ax.xaxis.tick_bottom()

    ax.set_ylabel('True', fontsize=20)
    ax.yaxis.set_ticklabels(class_names, fontsize = 10)
    plt.yticks(rotation=0)

    plt.title('Matrice de confusion', fontsize=20)

    plt.savefig('ConMat24.png')
    plt.show()

  def compare_result(test,pred):
    for i in range(0, len(test)):
        if pred[i] == test[i]:
            print ('Predicted: {0} --------- Actual:{1}'.format(pred[i], test[i]))
    for i in range(0, len(test)):
        if pred[i]!=test[i]:
            print ('Predicted: {0} +++++++++++++++ Actual:{1}'.format(pred[i], test[i]))

  def model_zip():
    pickle.dump(clf,open("/content/drive/MyDrive/repiceModel/recipe_model.sav", "wb"))
    

  def curve(clf,X,y):
    cv = StratifiedKFold(n_splits=3)
    visualizer = learning_curve(clf, X, y, cv=3, scoring='accuracy',n_jobs=20)
    visualizers = learning_curve(clf, X, y, cv=3, scoring='f1_weighted')
  

  
    
```
